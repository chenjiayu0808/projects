### Combining Data

```python
tommyjohn = pd.read_csv('tommyjohn.csv')
```

```python
tommyjohn.columns = tommyjohn.columns.str.strip()
tommyjohn.columns
```

```python
tommyjohn.head()
```

```python
for index, row in tommyjohn.iterrows():
    
    player = tommyjohn['name'][index]
    surgery_year = tommyjohn['year'][index]
    player_match = pstatsdf.loc[pstatsdf['Name'] == player]
    matches = player_match.loc[player_match['Year'] >= surgery_year]
    
    for idx in matches.index:
        
        pstatsdf['tommy_j'][idx] = 1
```

```python
pstatsdf.rename(columns={'tommy_j': 'past_surgery'}, inplace=True)
pstatsdf.sample(10)
```

```python
pstatsdf['injured'] = 0
pstatsdf.head()
```

```python
for index, row in DL_df.iterrows():
    
    player = DL_df['Player'][index]
    injury_year = DL_df['Year'][index]
    player_match = pstatsdf.loc[pstatsdf['Name'] == player]
    matches = player_match.loc[player_match['Year'] == injury_year]
    
    for idx in matches.index:
        
        pstatsdf['injured'][idx] = 1
```

```python
del pstatsdf['Tm']
del pstatsdf['Lg']
del pstatsdf['W']
del pstatsdf['L']
del pstatsdf['CG']
del pstatsdf['SHO']
del pstatsdf['SV']
del pstatsdf['H']
del pstatsdf['R']
del pstatsdf['ER']
del pstatsdf['HR']
del pstatsdf['IBB']
del pstatsdf['HBP']
del pstatsdf['BK']
del pstatsdf['WP']
del pstatsdf['HR9']
del pstatsdf['SO/W']
del pstatsdf['ERA+']
del pstatsdf ['FIP']
del pstatsdf['GS']
del pstatsdf['GF']
del pstatsdf['BB']
```

```python
pstatsdf = pstatsdf.dropna()
```

```python
pstatsdf.info()
```

```python
pstatsdf.rename(columns={'W-L%': 'win_percent', 'G': 'games_played', 'IP': 'innings_pitched', 'SO': 'strikeouts',
                         'BF': 'batters_faced', 'H9': 'h_ip', 'BB9':'bb_ip', 'SO9': 'so_ip'}, inplace=True)
```

```python
pstatsdf.info()
```

```python
ordered = ['Name', 'Age', 'Year', 'past_surgery', 'injured', 'Lefty', 'win_percent', 'ERA', 'games_played', 
           'innings_pitched', 'strikeouts', 'batters_faced', 'WHIP', 'h_ip', 'bb_ip', 'so_ip']
pstatsdf = pstatsdf[ordered]
```

```python
pstatsdf.injured.value_counts()
```

```python
group_player_by_year = pstatsdf.groupby(['Name','Year'])
statsdf = group_player_by_year.agg({'Age': 'mean', 'past_surgery':'mean', 'injured': 'mean', 'Lefty': 'mean',
                                    'win_percent': 'mean', 'ERA': 'mean', 'games_played': 'sum', 
                                    'innings_pitched': 'sum', 'strikeouts': 'sum', 'batters_faced': 'sum', 
                                    'WHIP': 'mean', 'h_ip': 'mean', 'bb_ip': 'mean', 'so_ip': 'mean'})
```

```python
statsdf.head(10)
```

```python
statsdf.reset_index(inplace=True)
```

```python
del statsdf['strikeouts']
del statsdf['WHIP']
```

```python
statsdf.head(10)
```

```python
statsdf.injured.value_counts()
```

```python
statsdf['avg_innings_per_game'] = statsdf['innings_pitched']/statsdf['games_played']
statsdf['avg_batters_per_game'] = statsdf['batters_faced']/statsdf['games_played']
del statsdf['innings_pitched']
del statsdf['batters_faced']
```

```python
statsdf.head(10)
```

```python
statsdf.to_csv("stats.csv")
```
