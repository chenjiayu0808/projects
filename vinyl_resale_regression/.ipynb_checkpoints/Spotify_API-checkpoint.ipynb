# Spotify API

```python
import requests
import urllib.parse

import seaborn as sns

import csv
import time
import random

import pandas as pd
import numpy as np
```

```python
vinyls = pd.read_pickle("./vinylsdf.pkl")
```

```python
artists = vinyls.Artist
artists.drop_duplicates(inplace=True)
artists.shape
```

```python
with open('artistpopularity8.csv', 'w',newline='') as csvfile:
    file = csv.writer(csvfile)

    file.writerow(['Name', 'Score'])
    
    for artist in artists:
        artist_encoded = urllib.parse.quote(artist)
        url = "https://api.spotify.com/v1/search?q=" + artist_encoded + "&type=artist&limit=1"
        r = requests.get(url, headers={"Accept": "application/json", "content-type":"application/json", 
                                   "Authorization": "Bearer"})
        json = r.json()
        
        try:
            name = json['artists']['items'][0]['name']
        except:
            name = 'Not Found'
            
        try:
            popularity_score = json['artists']['items'][0]['popularity']
        except:
            popularity_score = 'Not Found'
        
        observation = [name, popularity_score]
        
        file.writerow(observation)
    
        time.sleep(.5+2*random.random())
    
```

```python
artists = artists.reset_index(drop=True)
```

```python
artists.sample(10)
```

```python
df1 = pd.read_csv('artistpopularity.csv')
df2 = pd.read_csv('artistpopularity2.csv')
df3 = pd.read_csv('artistpopularity3.csv')
df4 = pd.read_csv('artistpopularity4.csv')
df5 = pd.read_csv('artistpopularity5.csv')
df6 = pd.read_csv('artistpopularity6.csv')
df7 = pd.read_csv('artistpopularity7.csv')
df8 = pd.read_csv('artistpopularity8.csv')
```

```python
spotifydf = df1.append(df2)
spotifydf = spotifydf.append(df3)
spotifydf = spotifydf.append(df4)
spotifydf = spotifydf.append(df5)
spotifydf = spotifydf.append(df6)
spotifydf = spotifydf.append(df7)
spotifydf = spotifydf.append(df8)
```

```python
spotifydf = spotifydf.reset_index(drop=True)
```

```python
spotifydf['Artists'] = artists
```

```python
spotifydf.sample(10)
```

```python
order = ['Artists', 'Score']
spotifydf = spotifydf[order]
spotifydf.sample(10)
```

```python
vinyls['Score'] = np.nan
```

```python
for index, row in spotifydf.iterrows():
    artist = spotifydf['Artists'][index]
    vinyls.loc[vinyls['Artist'] == artist, 'Score'] = spotifydf['Score'][index]
```

```python
spotvinyls = vinyls
vinylsfinal = spotvinyls[spotvinyls.Score != 'Not Found']
vinylsfinal.shape
```

```python
vinylsfinal['Score'] = pd.to_numeric(vinylsfinal.Score)
```
